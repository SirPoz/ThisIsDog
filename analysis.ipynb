# Dogbreed analysis

```python
import os

folders = os.listdir("Files")
folders
```

```python
dogs = []
for folder in folders:
    files = os.listdir(f'files/{folder}')
    dogs.append({"Folder":folder,"Count":len(files)})
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame.from_records(dogs)

plt.bar(df["Folder"],df["Count"])
plt.xticks(rotation=90)
plt.show()
```
